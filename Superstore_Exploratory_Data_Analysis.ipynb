```python
import ml_stats
```

```python
for col_name in ["Sales", "Quantity", "Discount", "Profit"]:
    print(col_name)
    stats_obj = ml_stats.ml_stats(df[col_name])
    stats_obj.stats_process()
    print("_"*100)
```

## Inference from above analysis: -

1) Sales: -
    * Range being so high(22638) means we the store has wide variety of products. They sell cheap products as well as expensive products.
    * Median being 54 and Mean 229 means there are OUTLIERS in our data. The difference is too much. Meaing people usually buy cheap products but some customers buy expensive products thus mean is very high.
    * Mode 12.96 means most of our customers buy 12.96 dollars product.
    * Q1 and Q3 represents majority of our customers buy products in the range of 17 to 209.
    * Q3 being 209 and mean(229) being greater than Q3 again proves we have too much OUTLIERS.
    * Total count of OUTLIERS is 1167 meaning out of 10000 people approx 1200 people are OUTLIERS that is 12% people. So if everyday 100 customers are coming, 88% will buy average products and 12% will buy expensive products.
    * Std being 623 and mean 229 meaning data points are 623 away from the mean. That means each customer has different buying habits. Majority of the customers does not show same buying habits.
    * 1 Standard Deviation range has 93.97% of data, which means majority of the business revolves around 1st Standard Deviation. Therefore if I analyse only 1st Standard Deviation, I'll understand 93.97% of the customer base.
    * Skewness being 12 meaning we have POSITIVE OUTLIERS. Our data is Right Skewed.
    * Kurtosis being 305 means the data has a sharp peak around the mean. Which means majority of the customers buy products around the mean.
    * 95% confidence interval is between 217 to 242 which means we are 95% confidence that our best guess at truth lies between 217 to 242. The mean(229) is the average of this sample of 10000 customers but the superstore serves more than 10k customers. Therefore we use inferential statistics to predict for the population and we use confidence interval to give the best range.
